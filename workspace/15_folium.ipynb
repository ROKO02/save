```python
# folium은 Leaflet.js를 이용해서 위치 좌표계 값을 지도위에 시각화 할 수 있다
# !pip install folium
import folium
# 보고자하는 위치의 좌표계를 얻어내는 가장 단순한 방법은 구글맵을 이용하는 것이다
```

```python
# folium 라이브러리의 Map() 함수의 location 속성으로 지정한 위치를 중심으로 하는 zoom_start 옵션으로 지정한 배율을 사용하는
# 지도를 만든다
# folium.Map(location=[위도, 경도], zoom_start=배율, tiles='지도 스타일')
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=17)

# folium의 save() 함수로 작성된 지도를 html 파일로 저장할 수 있다
map_folium.save('./output/map.html')
map_folium
```

```python
# 지도 스타일 : Open Street Map(기본값), Stamen Terrain, Stamen Toner
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=17, tiles='Open Street Map')
map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=17, tiles='Stamen Terrain')
map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=17, tiles='Stamen Toner')
map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=16)

# folium 라이브러리의 Marker() 함수의 location 속성으로 지정한 위치에 마커를 만든다
# popup 속성으로 마커를 클릭했을때 표시할 문자열을 지정할 수 있다
# 작성된 마커는 add_to() 함수로 지도에 추가한다
folium.Marker(location=[37.56648, 126.97792], popup='보신각').add_to(map_folium)
folium.Marker(location=[37.56999, 126.98731], popup='서울특별시청').add_to(map_folium)
folium.Marker(location=[37.565849255025526, 126.97519831679298], popup='경복궁').add_to(map_folium)
folium.Marker(location=[37.57631746728676, 126.9767564579018], popup='광화문').add_to(map_folium)

map_folium.save('./output/map.html')
map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=16)

# 마커를 클릭했을때 팝업되는 말풍선의 크기가 마음에 들지 않으면 folium의 Popup() 함수의 max_width 속성으로 변경하면 된다
popup = folium.Popup('보신각', max_width=100)

folium.Marker(location=[37.56648, 126.97792], popup=popup).add_to(map_folium)

map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=16)

# icon 속성에 folium의 Icon() 함수로 마커의 색상과 아이콘의 모양을 지정할 수 있다
# Icon() 함수의 color 속성으로 지정할 수 있는 색상은 아래와 같다
# white', 'lightred', 'green', 'pink', 'orange', 'beige', 'darkred', 'darkgreen', 'cadetblue', 'darkpurple', 'lightgray', 
#'purple', 'black', 'blue', 'red', 'lightblue', 'gray', 'darkblue', 'lightgreen'
# 마커에 표시되는 아이콘은 https://getbootstrap.com/docs/3.3/components/#glyphicons-glyph에서 제공하는 이미지를 사용한다

folium.Marker(
    location=[37.56648, 126.97792], 
    popup=folium.Popup('보신각', max_width=100),
    icon=folium.Icon(color='pink', icon='star')
).add_to(map_folium)
folium.Marker(
    location=[37.57631746728676, 126.9767564579018], 
    popup=folium.Popup('광화문', max_width=100),
    icon=folium.Icon(color='black', icon='hand-down')
).add_to(map_folium)
folium.Marker(
    location=[37.565849255025526, 126.97519831679298], 
    popup=folium.Popup('경복궁', max_width=100),
    icon=folium.Icon(color='beige', icon='tower')
).add_to(map_folium)

map_folium
```

```python
map_folium = folium.Map(location=[37.56999, 126.98731], zoom_start=16)

# folium의 CircleMarker() 함수로 원 모양의 마커를 만들 수 있다
folium.CircleMarker(
    location=[37.56648, 126.97792], 
    popup=folium.Popup('보신각', max_width=100),
    radius=100, color='#16EB64', fill_color='#CAE838'
).add_to(map_folium)

map_folium
```
