```python
import random
```

```python
# 1
for i in range(1,10)[:3]:
    print('=' * 40)
    fc, vc = map(int,input('고정비용, 가변비용 입력 : ').split())
    num_product = int(input('제품 생산개수 : '));
    cost_total = fc + (vc * num_product)
    print('총 드는 비용 : {}'.format(cost_total))

    while True:
        price = int(input('제품 판매 가격 : '))
        if price >= vc:
            bep = cost_total // price + 1
            print('손익분기점 : {}개'.format(bep))
            net_profits = (price * num_product) - cost_total
            print('최대 얻을 수 있는 순이익 : {}만원'.format(net_profits))
            break
        
        else:
            print('손익분기점이 없습니다 \n다시 입력하세요')
        
    customer = 0
    product = 0
    netprofit = 0
    for j in range(1,31):
        print('=' * 22)
        customers = random.randint(1,100)
    
        while True:
            sell_products = random.randint(1, num_product)
            if sell_products > num_product:
                sell_products = random.randint(1, num_product)
            else:
                break
            
        day_profits = price * sell_products
        day_Netprofits = day_profits - cost_total
    
        print('▶'* 4 + '{}일째'.format(j) + '◀' * 4)
        print('방문 고객 수 : {}명'.format(customers))
        print('제품 판개 개수 : {}개'.format(sell_products))
        print('판매 이익 : {}만원'.format(day_profits))
        print('순이익 : {}만원'.format(day_Netprofits))
        customer += customers
        product += sell_products
        netprofit += day_Netprofits
        
    if netprofit >= 0: 
        hun_mil = netprofit // 10000
        ten_tho = netprofit % 10000
        print('=' * 24)
        print('{}달 총 고객 수 : {}명\n총 제품 판매수 : {}개'.format(i,customer,product))
        # print(netprofit)
        if hun_mil == 0:
            print('총 순이익 : {}만원'.format(netprofit))
        else:
            if ten_tho == 0:
                print('총 순이익 : {}억원'.format(hun_mil))
            else:
                print('총 순이익 : {}억 {}만원'.format(hun_mil, ten_tho))
    elif netprofit < 0:
        hun_mil = -(netprofit // -10000)
        ten_tho = netprofit % -10000
        print('=' * 24)
        print('{}달 총 고객 수 : {}명\n총 제품 판매수 : {}개'.format(i,customer,product))
        # print(netprofit)
        if hun_mil == 0:
            print('총 순이익 : {}만원'.format(netprofit))
        else:
            if ten_tho == 0:
                print('총 순이익 : {}억원'.format(hun_mil))
            else:
                print('총 순이익 : {}억 {}만원'.format(hun_mil, ten_tho))
    
    if hun_mil <= -1:
        print('파산하셨습니다')
        break
    
```
